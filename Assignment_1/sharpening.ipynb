#### Questions 1-2
*sharpened image = original image + details*

*details = original image - blurred image*



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

IMG_PATH_RICE = 'rice.png'
img_rice = cv2.imread(IMG_PATH_RICE)

img_rice_bblur = cv2.blur(src=img_rice, ksize=(5,5))
img_rice_gblur = cv2.GaussianBlur(src=img_rice, 
                                  ksize=(5,5), 
                                  sigmaX=np.sqrt(2), 
                                  sigmaY=np.sqrt(2))

img_rice_bsharp = 2 * img_rice - img_rice_bblur
img_rice_gsharp = 2 * img_rice - img_rice_gblur

plt.rcParams['figure.figsize'] = (15,10)

std_box_sharp = cv2.meanStdDev(img_rice_bsharp)
std_gauss_sharp = cv2.meanStdDev(img_rice_gsharp)

plt.subplot(131), plt.imshow(img_rice)
plt.xticks([]), plt.yticks([]), plt.title('Figure 1(a) Original')

plt.subplot(132), plt.imshow(img_rice_bsharp), plt.title('Figure 1(b) Sharpened with Box')
plt.xticks([]), plt.yticks([])
plt.xlabel(str.format('PSNR: {}', cv2.PSNR(img_rice, img_rice_bsharp)))

plt.subplot(133), plt.imshow(img_rice_gsharp), plt.title('Figure 1(c) Sharpened with Gaussian')
plt.xticks([]), plt.yticks([])
plt.xlabel(str.format('PSNR: {}', cv2.PSNR(img_rice, img_rice_gsharp)))

plt.show()
```

By comparing the PSNR of *Figure 1(b)* and *Figure 1(c)*, one can see that *Figure 1(c)* yields a higher PSNR. Therefore, the Gaussian Filter is more suited for sharpening. This would make sense because the Gaussian Filter is a better low pass filter than the Box filter. Hence, the subtraction between 2x the original image and the Gaussian filtered image passes higher higher values than the subtraction with the Box filter. The smoother the image, the more details get highlighted, yielding a sharper image. Once again proving that the Gaussian Filter is more suited for sharpening use as opposed to the Box Filter.

```python
img_rice_lap = cv2.Laplacian(src=img_rice, ddepth=cv2.CV_8U, ksize=3)
img_rice_blap = cv2.Laplacian(src=img_rice_bsharp, ddepth=cv2.CV_8U, ksize=3)
img_rice_glap = cv2.Laplacian(src=img_rice_gsharp, ddepth=cv2.CV_8U, ksize=3)

diff_1 = img_rice_blap - img_rice_glap
diff_2 = img_rice_glap - img_rice_blap

plt.rcParams['figure.figsize'] = (15,10)
plt.subplot(131), plt.imshow(img_rice_lap)
plt.xticks([]), plt.yticks([]), plt.title('Figure 2(a) Laplacian of Original Images')

plt.subplot(132), plt.imshow(diff_1)
plt.xticks([]), plt.yticks([]), plt.title('Figure 2(b) Box Edges - Gaussian Edges')

plt.subplot(133), plt.imshow(diff_2)
plt.xticks([]), plt.yticks([]), plt.title('Figure 2(c) Gaussian Edges - Box Edges')

plt.show()
```

It can also be noted by comparing *Figure 2(b)* and *Figure 2(c)* that subtracting the Box sharpened edges by the Gaussian sharpened image yields no edges while as substracting the Gaussian sharpened edges by the Box sharpened edges still yields edges. This means that the Box's edge values are weaker than the Gaussian's edge values, proving that the Gaussian method for sharpening is more effective.
