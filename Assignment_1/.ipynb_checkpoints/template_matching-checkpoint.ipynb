# 3. Template Matching
#### Question 1-2

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

IMG_PATH_PARK = 'where_is_waldo.jpg'
IMG_PATH_TEMP1 = 'template1.png'
IMG_PATH_TEMP2 = 'template2.png'

img_park = cv2.imread(IMG_PATH_PARK)
img_temp = cv2.imread(IMG_PATH_TEMP1)
img_temp_noise = cv2.imread(IMG_PATH_TEMP2)

# Have to convert BGR to RGB otherwise the colors are in BGR, which doesn't match with the input
img_park = cv2.cvtColor(img_park, cv2.COLOR_BGR2RGB)
img_temp = cv2.cvtColor(img_temp, cv2.COLOR_BGR2RGB)
img_temp_noise = cv2.cvtColor(img_temp_noise, cv2.COLOR_BGR2RGB)

temp_ht, temp_wt = img_temp.shape[0:2]

# Compute Template Matching by using the Sum of Square Difference method
tm_res = cv2.matchTemplate(img_park, img_temp, cv2.TM_SQDIFF)
tm_res_noise = cv2.matchTemplate(img_park, img_temp_noise, cv2.TM_SQDIFF)

img_res = img_park.copy()
img_res_noise = img_park.copy()

# These thresholds are used to determine the minimum difference
THRESH = 8
THRESH_NOISE = 1_973_689

# Return back values which are under the threshold
loc = np.where(tm_res <= THRESH)
pt_min = []
# Have to iterate over the values in reverse because they are in a y,x convention
# cv2.rectangle requires a x,y convention
for pt in zip(*loc[::-1]):
    cv2.rectangle(img_res, pt, (pt[0] + temp_ht, pt[1] + temp_wt), (0, 255, 0), 10)
    pt_min = [pt[0], pt[1]]

loc_noise = np.where(tm_res_noise <= THRESH_NOISE)
for pt in zip(*loc_noise[::-1]):
    cv2.rectangle(img_res_noise, pt, (pt[0] + temp_ht, pt[1] + temp_wt), (0, 255, 0), 10)
    
# Make the figures large because the rectangle match is small and difficult to see
plt.figure(figsize=(20, 20))
plt.subplot(321), plt.imshow(img_park)
plt.xticks([]), plt.yticks([]), plt.title('Figure 1(a) Original Image')
plt.subplot(323), plt.imshow(img_temp)
plt.xticks([]), plt.yticks([]), plt.title('Figure 1(b) Template Image')
plt.subplot(324), plt.imshow(img_temp_noise)
plt.xticks([]), plt.yticks([]), plt.title('Figure 1(c) Noisy Template Image')
plt.subplot(325), plt.imshow(img_res)
plt.xticks([]), plt.yticks([]), plt.title('Figure 1(d) Matched Image'), plt.xlabel(str.format('Threshold: {}\n(x,y) of Min SSD Value: {}', THRESH, pt_min))
plt.subplot(326), plt.imshow(img_res_noise)
plt.xticks([]), plt.yticks([]), plt.title('Figure 1(e) Matched Image with Noisy Template'), plt.xlabel(str.format('Threshold: {}\n(x,y) of Min SSD Value: {}', THRESH_NOISE, pt_min))
plt.tight_layout()
plt.show()
```
