# 2. Edge Detection
## 2.2 Laplacian of Gaussian
#### Question 1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

IMG_WOMAN_PATH = 'woman.png'
img_woman = cv2.imread(IMG_WOMAN_PATH)

woman_gauss = cv2.GaussianBlur(img_woman, (5,5), np.sqrt(2), np.sqrt(2))
woman_gauss = cv2.threshold(woman_gauss, 150, 255, cv2.THRESH_BINARY)[1]
woman_lapgauss = cv2.Laplacian(woman_gauss, cv2.CV_8U, ksize=5)

plt.rcParams['figure.figsize'] = (20,10)
plt.subplot(121), plt.imshow(img_woman)
plt.xticks([]), plt.yticks([]), plt.title('Figure 1(a) Original')
plt.subplot(122), plt.imshow(woman_lapgauss)
plt.xticks([]), plt.yticks([]), plt.title('Figure 1(b) Laplacian of Gaussian Edge Detector')
plt.show()
```
