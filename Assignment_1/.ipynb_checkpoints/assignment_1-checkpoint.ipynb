# 1 Filtering
## 1.1 Denoising

```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt

def psnr(imageA, imageB) :
    """Compute the Peak Signal to Noise Ratio between the two images.
    NOTE: The two images must be of the same dimension.
    Formula taken from https://www.mathworks.com/help/vision/ref/psnr.html
    """    
    if (imageA.shape != imageB.shape) : 
        return -1
    
    # Compute the MSE
    mse = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    mse /= float(imageA.shape[0] * imageA.shape[1])
    
    R = 2 ** (imageA.itemsize * 8)
    return 10 * math.log10((R ** 2) / mse)

img_path_1 = 'peppers.png'
img_path_2 = 'gaussian_noise_peppers.png'
img_path_3 = 'saltandpepper_noise_peppers.png'

img_original = cv2.imread(img_path_1)
img_noisy_gauss = cv2.imread(img_path_2)

# Standard Deviation = sqrt(Variance)
res_gauss = cv2.GaussianBlur(src=img_noisy_gauss, 
                             ksize=(7,7), 
                             sigmaX=math.sqrt(2), 
                             sigmaY=math.sqrt(2))

# Box filter requires a matrix to compute the average of n neighboring pixels
M = np.ones((7,7), np.float32)/49
res_box = cv2.filter2D(src=img_noisy_gauss, ddepth=-1, kernel=M)

psnr_gauss = psnr(img_original, res_gauss)
psnr_box = psnr(img_original, res_box)

plt.rcParams['figure.figsize'] = (15,10)
plt.subplot(221), plt.imshow(img_original), plt.title('Figure 1(a) Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(img_noisy_gauss), plt.title('Figure 1(b) Noisy Image')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(res_gauss), plt.title('Figure 1(c) Gaussian Filtering')
plt.xticks([]), plt.yticks([]), plt.xlabel(str.format("PSNR = {}", psnr_gauss))
plt.subplot(224), plt.imshow(res_box), plt.title('Figure 1(d) Box Filtering')
plt.xticks([]), plt.yticks([]), plt.xlabel(str.format("PSNR = {}", psnr_box))
plt.show()
```

As stated under *mathworks.com/help/vision/ref/psnr.html* The PSNR 'ratio is often used as a quality measurement between the original and a compressed image. __The higher the PSNR, the better the quality__ of the compressed, or reconstructed image.'

One can compare the PSNR of Figures 1(c) and 1(d) to note that *Figure 1(c)*, the *Gaussian Filtering* leads to a __higher PSNR__, thus a better quality of reconstruction. One can also see how Figure 1(c) ressembles more closely to Figure 1(a) than Figure 1(d) ressembling to Figure 1(a).

```python
img_noisy_pepper = cv2.imread(img_path_3)

res_gauss = cv2.filter2D(src=img_noisy_pepper, ddepth=-1, kernel=M)
res_median = cv2.medianBlur(src=img_noisy_pepper, ksize=7)

psnr_gauss = psnr(img_original, res_gauss)
psnr_median = psnr(img_original, res_median)

plt.rcParams['figure.figsize'] = (15,10)
plt.subplot(221), plt.imshow(img_original), plt.title('Figure 2(a) Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(img_noisy_pepper), plt.title('Figure 2(b) Salt and Pepper Noise')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(res_gauss), plt.title('Figure 2(c) Gaussian Filtering')
plt.xticks([]), plt.yticks([]), plt.xlabel(str.format("PSNR = {}", psnr_gauss))
plt.subplot(224), plt.imshow(res_median), plt.title('Figure 2(d) Median Filtering')
plt.xticks([]), plt.yticks([]), plt.xlabel(str.format("PSNR = {}", psnr_median))
plt.show()
```

As one can see by comparing Figures 2(c) and 2(d), the *Median Filtering* results in a better reconstructed image. The __*PSNR* of *Figure 2(d)* is larger than that of *Figure 2(c)*.__ The PSNR of the *Median Filter* is of roughly 27.13 while as that of the *Gaussian Filtering* is of roughly 22.71. This proves that *Median Filtering* is more suitable for a *Salt and Pepper* noisy image reconstruction.

## 1.2 Sharpening

*sharpened image = original image + details*

*details = original image - blurred image*

```python
img_path_4 = 'rice.png'
img_rice = cv2.imread(img_path_4)

M = np.ones((5,5), np.float32)/25
img_rice_bblur = cv2.filter2D(src=img_rice, ddepth=cv2.CV_8U, kernel=M)
img_rice_gblur = cv2.GaussianBlur(src=img_rice, 
                                  ksize=(5,5), 
                                  sigmaX=math.sqrt(2), 
                                  sigmaY=math.sqrt(2))

img_rice_bsharp = 2 * img_rice - img_rice_bblur
img_rice_gsharp = 2 * img_rice - img_rice_gblur

plt.rcParams['figure.figsize'] = (15,10)
plt.subplot(131), plt.imshow(img_rice)
plt.xticks([]), plt.yticks([]), plt.title('Figure 3(a) Original')
plt.subplot(132), plt.imshow(img_rice_bsharp), plt.title('Figure 3(b) Sharpened with Box')
plt.xticks([]), plt.yticks([]), plt.xlabel(str.format('PSNR {}', psnr(img_rice, img_rice_bsharp)))
plt.subplot(133), plt.imshow(img_rice_gsharp), plt.title('Figure 3(c) Sharpened with Gaussian')
plt.xticks([]), plt.yticks([]), plt.xlabel(str.format('PSNR {}', psnr(img_rice, img_rice_gsharp)))
plt.show()
```

```python
# Determine the 'sharpness' of an image by the use of a Laplacian kernel
M = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])

plt.rcParams['figure.figsize'] = (15,10)
img_rice_blap = cv2.filter2D(src=img_rice_bsharp, ddepth=cv2.CV_8U, kernel=M)
img_rice_glap = cv2.filter2D(src=img_rice_gsharp, ddepth=cv2.CV_8U, kernel=M)
plt.subplot(121), plt.imshow(img_rice_blap)
plt.xticks([]), plt.yticks([]), plt.title('Figure 4(a) Edges of Box Sharpened Image')
plt.subplot(122), plt.imshow(img_rice_glap)
plt.xticks([]), plt.yticks([]), plt.title('Figure 4(b) Edges of Gaussian Sharpened Image')
plt.show()
```

#### 2. Effect of denoising on edge detection

```python
#TODO verify threshold. edges don't look right

img_path_woman = 'woman.png'
img_woman = cv2.imread(img_path_woman)

# Direct Sobel filter on woman
woman_sobel = np.uint8(np.hypot(cv2.Sobel(img_woman, cv2.CV_64F, 1, 0, 3), cv2.Sobel(img_woman, cv2.CV_64F, 0, 1, 3)))

# Sobel filter with denoising using a box filter
M = np.ones((3,3))/9
woman_box = cv2.filter2D(img_woman, -1, M)
woman_box_sobel = np.uint8(np.hypot(cv2.Sobel(woman_box, cv2.CV_64F, 1, 0, 3), cv2.Sobel(woman_box, cv2.CV_64F, 0, 1, 3)))

ws_thresh10 = cv2.threshold(woman_sobel, 0.1 * np.max(woman_sobel), 255, cv2.THRESH_BINARY)[1]
ws_thresh20 = cv2.threshold(woman_sobel, 0.2 * np.max(woman_sobel), 255, cv2.THRESH_BINARY)[1]

ws_box_thresh10 = cv2.threshold(woman_box_sobel, 0.1 * np.max(woman_box_sobel), 255, cv2.THRESH_BINARY)[1]
ws_box_thresh20 = cv2.threshold(woman_box_sobel, 0.2 * np.max(woman_box_sobel), 255, cv2.THRESH_BINARY)[1]

plt.subplot(241), plt.imshow(img_woman)
plt.xticks([]), plt.yticks([]), plt.title('Figure 6(a) Original')
plt.subplot(242), plt.imshow(ws_thresh10)
plt.xticks([]), plt.yticks([]), plt.title('Figure 6(b) Sobel Edge Detector without Denoising\n at a Threshold of 10% of Maximum Value')
plt.subplot(243), plt.imshow(ws_thresh20)
plt.xticks([]), plt.yticks([]), plt.title('Figure 6(c) Sobel Edge Detector without Denoising\n at a Threshold of 20% of Maximum Value')
plt.subplot(244), plt.imshow(ws_box_thresh10)
plt.xticks([]), plt.yticks([]), plt.title('Figure 6(d) Sobel Edge Detector with Denoising\n at a Threshold of 10% of Maximum Value')
plt.subplot(245), plt.imshow(ws_box_thresh20)
plt.xticks([]), plt.yticks([]), plt.title('Figure 6(e) Sobel Edge Detector with Denoising\n at a Threshold of 20% of Maximum Value')
plt.tight_layout()
plt.show()
```
