```python
def computeEuclidean(a, b, ax=1):
    """
    @brief Compute euclidean distance between each vector within lists a, and b
    @param a List of vectors
    @param b List of vectors
    @ax Axis to project distance on
    """
    return np.linalg.norm(a - b, axis=ax)
```

```python
def computeBayes(point, mean, std):
    return (1 / (2 * np.sqrt(2 * np.pi) * std)) * np.exp(-1 * ((point - mean) ** 2) / (2 * std ** 2))
```

#### Question 1.1.1 Implement the K-means clustering algorithm

```python
def computeKMeans(img, k_clusters=3, n_iterations=10):
    """
    @brief Categorize each pixel of the given image to a cluster label
    @param img The source image to compute the clusters for
    @param k_clusters The number of clusters to compute
    @param n_iterations The number of iterations to iterate for cluster convergence
    """
    MIN_RGB_VAL = 0
    MAX_RGB_VAL = 255
    height, width = img.shape[0:2]
    result = np.copy(img)

    # An initial state for the centroids and values to be clustered must be defined manually
    C = [np.random.randint(MIN_RGB_VAL, MAX_RGB_VAL, size=3, dtype=np.uint8) for k in range(0, k_clusters)]
    clusters = np.zeros(img.shape[0:2])
    
    for iteration in range(n_iterations):
                
        for x in range(0, height):
            for y in range(0, width):
                
                rgb = img[x,y]
                
                distances = computeEuclidean(rgb, C)
                closest_centroid = np.argmin(distances)        
                clusters[x, y] = closest_centroid
        
        for i in range(k_clusters):
            points = []
            for x in range(0, height):
                for y in range(0, width):
                    if (clusters[x,y] == i):
                        rgb = img[x,y]
                        points.append(rgb)
            C[i] = np.mean(np.array(points), axis=0)
    
    for x in range(0, height):
        for y in range(0, width):
            cluster = np.uint8(clusters[x,y])
            result[x,y] = C[cluster]
        
    return result
```

Compute K-means clustering to the provided image with K=2 and K=3 over 10 iterations

```python
kmeans_3 = computeKMeans(img_home, k_clusters=3)
kmeans_2 = computeKMeans(img_home, k_clusters=2)
```

Display the segmented images

```python
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(kmeans_3)
plt.xticks([]), plt.yticks([]), plt.title('K=3 KMeans Cluster Segmentation on home.jpg')
plt.subplot(122)
plt.imshow(kmeans_2)
plt.xticks([]), plt.yticks([]), plt.title('K=2 KMeans Cluster Segmentation on home.jpg')
plt.tight_layout()
plt.show()
```
